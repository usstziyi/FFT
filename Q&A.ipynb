# 为什么令 $\omega_k = \frac{2\pi k}{N}$

核心原因：**保证 N 个基向量在 N 维空间中两两正交**。

## 从头推导

我们有 $N$ 个离散样本 $x[0], x[1], \dots, x[N-1]$，想把它投影到一组复指数基函数上：

$$x[n] = \sum_k X[k] \cdot e^{j\omega_k n}, \quad n = 0,1,\dots,N-1$$

每个基向量是 $N$ 维空间中的一个向量：

$$\mathbf{v}_k = \begin{pmatrix} e^{j\omega_k \cdot 0} \\ e^{j\omega_k \cdot 1} \\ \vdots \\ e^{j\omega_k \cdot (N-1)} \end{pmatrix}$$

## 正交性条件

要让投影能完美分解，$\mathbf{v}_k$ 之间必须**两两正交**：

$$\langle \mathbf{v}_k, \mathbf{v}_m \rangle = \sum_{n=0}^{N-1} e^{j(\omega_k - \omega_m) \cdot n} = 0 \quad (k \neq m)$$

这是一个**等比数列求和**：

$$\sum_{n=0}^{N-1} e^{j\Delta\omega \cdot n} = \frac{1 - e^{j\Delta\omega \cdot N}}{1 - e^{j\Delta\omega}}$$

要让它为零，**分子必须为零**：

$$e^{j\Delta\omega \cdot N} = 1 \implies \Delta\omega \cdot N = 2\pi \cdot m \quad (m \in \mathbb{Z})$$

即：

$$\boxed{\Delta\omega = \frac{2\pi m}{N}}$$

## 为什么恰好选 $\omega_k = \frac{2\pi k}{N}$

我们需要满足**两个约束**：

| 约束 | 要求 | $\omega_k = \frac{2\pi k}{N}$ 是否满足 |
|---|---|---|
| 正交性 | 任意 $\Delta\omega$ 是 $\frac{2\pi}{N}$ 的整数倍 | $\omega_k - \omega_m = \frac{2\pi(k-m)}{N}$，$k-m$ 是整数，满足 |
| 数量 | 恰好 $N$ 个线性无关的基向量 | $k=0,1,\dots,N-1$ 给出 $N$ 个不同的基向量，刚好撑满 $N$ 维空间 |

如果频率间隔**小于** $\frac{2\pi}{N}$，基向量不够 $N$ 个，无法完整表示信号；如果**大于** $\frac{2\pi}{N}$，又会丢掉信息。

所以 $\omega_k = \frac{2\pi k}{N}$ 是**唯一**能同时满足正交性和完备性的选择。

## 直观理解

把 $N$ 个样本看作 $N$ 维空间中的一个点。我们需要找 $N$ 个"方向"来分解它。每个复指数 $e^{j\omega_k n}$ 就是一个方向。$\omega_k = \frac{2\pi k}{N}$ 的含义是：

> 第 $k$ 个基函数在 $N$ 个采样点上恰好走过 $k$ 个完整周期。

- $k=0$：常数（直流分量）
- $k=1$：走 1 个周期
- $k=2$：走 2 个周期
- ...
- $k=N/2$：走 $N/2$ 个周期（Nyquist 频率，每个采样点交替正负）

每多走一个周期，基函数就和前面所有的基函数正交——就像三维空间中 $x, y, z$ 轴互相垂直一样。

因为你这里频域右边那个量**不是直接的“每个频点功率”本身**，而是 **DFT 系数的平方和**。
而在你采用的这套 DFT 定义下，(X[k]) 里天然多带了一个 **(N) 倍的尺度**，所以要补一个 ($\frac{1}{N}$) 才能和时域能量对应起来。

你图里的式子是帕塞瓦尔关系在这套 DFT 定义下的形式：

$$\sum_{n=0}^{N-1}|x[n]|^2=\frac{1}{N}\sum_{k=0}^{N-1}|X[k]|^2$$

---

## 1. 先说结论

频域右边之所以要除以 (N)，**不是因为频域天生就该除以 (N)**，而是因为你定义的 DFT 是：

$$
X[k]=\sum_{n=0}^{N-1}x[n]e^{-j2\pi kn/N}
$$

$$
x[n]=\frac{1}{N}\sum_{k=0}^{N-1}X[k]e^{j2\pi kn/N}
$$

也就是说：

* 正变换 **不除 (N)**
* 逆变换 **除 (N)**

在这种定义下，频域系数 (X[k]) 的量级会比“真正对应的能量系数”大一些，所以平方求和后会多出一个 (N)，于是必须再乘 ($\frac{1}{N}$)。

---

## 2. 为什么会多出一个 (N)

关键在于 DFT 基函数的“长度”不是 1，而是 ($\sqrt{N}$) 这个量级。

离散复指数基函数是：

$$
\phi_k[n]=e^{j2\pi kn/N}
$$

它的模长恒为 1，所以在 (n=0,\dots,N-1) 上，

$$
\sum_{n=0}^{N-1}|\phi_k[n]|^2 = N
$$

这说明这些基函数虽然彼此正交，但**没有归一化到单位长度**。
它们的范数平方是 (N)，不是 1。

所以当你拿 (x[n]) 去投影时，得到的 (X[k]) 实际上是“带着基函数长度因子”的系数。

---

## 3. 用向量投影的角度看

如果一组正交基 ({u_k}) 不是单位正交基，而是满足

$$
\langle u_k,u_m\rangle =
\begin{cases}
N, & k=m\
0, & k\neq m
\end{cases}
$$

那么信号展开系数不能直接写成 (\langle x,u_k\rangle)，而要写成

$$
c_k=\frac{\langle x,u_k\rangle}{\langle u_k,u_k\rangle} = \frac{1}{N}\langle x,u_k\rangle
$$

而 DFT 里的 (X[k]) 实际上就是那个内积：

$$
X[k]=\sum_{n=0}^{N-1}x[n]e^{-j2\pi kn/N}
$$

所以真正展开系数其实是

$$
c_k=\frac{X[k]}{N}
$$

也就是说，(X[k]) 比真正的展开系数大了 (N) 倍。

那能量怎么算？
按正交展开，能量应该是“系数平方 × 基函数能量”：

$$
\sum_{n=0}^{N-1}|x[n]|^2 = \sum_{k=0}^{N-1}|c_k|^2\cdot N
$$

把 ($c_k=X[k]/N$) 代进去：

$$
\sum_{n=0}^{N-1}|x[n]|^2 = \sum_{k=0}^{N-1}\left|\frac{X[k]}{N}\right|^2 N = \frac{1}{N}\sum_{k=0}^{N-1}|X[k]|^2
$$

这就是为什么会出现 ($\frac{1}{N}$)。

---

## 4. 你可以把它理解成“DFT 系数平方和太大了”

因为 (X[k]) 本身是一个长度为 (N) 的求和：

$$
X[k]=\sum_{n=0}^{N-1}x[n]e^{-j2\pi kn/N}
$$

每个频点都是把 (N) 项累加起来的，所以它天然就带有一个和 (N) 相关的放大量。

因此：

* 时域直接平方求和：($\sum |x[n]|^2$)
* 频域如果直接平方求和：($\sum |X[k]|^2$)

后者会比前者大 (N) 倍，于是要除以 (N)。

---

## 5. 用一个最简单的例子验证

取一个常数序列：

$$
x[n]=1,\quad n=0,\dots,N-1
$$

### 时域能量

$$
\sum_{n=0}^{N-1}|x[n]|^2 = \sum_{n=0}^{N-1}1 = N
$$

### 频域 DFT

只有 (k=0) 处非零：

$$
X[0]=N,\qquad X[k]=0\ (k\neq 0)
$$

所以频域平方和是

$$
\sum_{k=0}^{N-1}|X[k]|^2 = |N|^2 = N^2
$$

如果你不除以 (N)，就得到 (N^2)，和时域的 (N) 对不上。
除以 (N) 后：

$$
\frac{1}{N}\sum_{k=0}^{N-1}|X[k]|^2 = \frac{1}{N}N^2 = N
$$

刚好和时域能量一致。

---

## 6. 这不是“频域总功率要除以 N”，而是“这套归一化下要除以 N”

这个点很重要。

如果你换一种 DFT 定义，例如对称归一化：

$$
X[k]=\frac{1}{\sqrt{N}}\sum_{n=0}^{N-1}x[n]e^{-j2\pi kn/N}
$$

$$
x[n]=\frac{1}{\sqrt{N}}\sum_{k=0}^{N-1}X[k]e^{j2\pi kn/N}
$$

那么帕塞瓦尔关系就会变成

$$
\sum_{n=0}^{N-1}|x[n]|^2 = \sum_{k=0}^{N-1}|X[k]|^2
$$

这时就**不用再除 (N)** 了。

所以本质不是频域有什么神秘的“除 (N) 规律”，而是：

**你选的 DFT 归一化方式决定了帕塞瓦尔公式右边前面的系数。**

---

## 7. 能量和平均功率别混了

你这张图写的是：

**时域能量 = 频域能量**

所以对应的是总能量：

$$
\sum |x[n]|^2
$$

如果你想写平均功率，还要再除以 (N)：

$$
P_{avg}=\frac{1}{N}\sum_{n=0}^{N-1}|x[n]|^2
$$

再结合帕塞瓦尔：

$$
P_{avg} = \frac{1}{N}\cdot \frac{1}{N}\sum_{k=0}^{N-1}|X[k]|^2 = \frac{1}{N^2}\sum_{k=0}^{N-1}|X[k]|^2
$$

所以要分清：

* **总能量**：($\sum |x[n]|^2$)
* **平均功率**：($\frac{1}{N}\sum |x[n]|^2$)

在你这套 DFT 下：

* 时域总能量 ↔ ($\frac{1}{N}\sum |X[k]|^2$)
* 时域平均功率 ↔ ($\frac{1}{N^2}\sum |X[k]|^2$)

---

## 8. 一句话总结

**因为你定义的 DFT 系数 (X[k]) 没有做单位化，平方和会比时域能量大 (N) 倍，所以频域总能量要乘 ($\frac{1}{N}$) 才能和时域对应。**

如果你愿意，我下一步可以继续给你严格推导一遍这个公式，从

$$
x[n]=\frac{1}{N}\sum_{k=0}^{N-1}X[k]e^{j2\pi kn/N}
$$

一步一步推到

$$
\sum |x[n]|^2=\frac{1}{N}\sum |X[k]|^2
$$

把中间每一步正交性怎么用都写出来。
